```python
import pandas as pd
```

```python
secs=pd.read_csv('W_HSS.txt',delimiter="\t").set_index('AISC_Manual_Label')
secs
```

```python
#
cs=pd.DataFrame({'connection_ID':[],'primary_ele':[],'secondary_ele':[],'angle':[]}).set_index('connection_ID')
num=0
for sec1 in secs.index:
    for sec2 in secs.index:
        if sec1 is not sec2:
            num=num+1
            connection_id='AA'+str(num)+'00'
            cs_new=pd.DataFrame({'connection_ID':connection_id,'primary_ele':sec1,'secondary_ele':sec2,'angle':0},index=['connection_ID']).set_index('connection_ID')
            cs=cs.append(cs_new)
            connection_id='AA'+str(num)+'45'
            cs_new=pd.DataFrame({'connection_ID':connection_id,'primary_ele':sec1,'secondary_ele':sec2,'angle':45},index=['connection_ID']).set_index('connection_ID')
            cs=cs.append(cs_new)
            connection_id='AA'+str(num)+'90'
            cs_new=pd.DataFrame({'connection_ID':connection_id,'primary_ele':sec1,'secondary_ele':sec2,'angle':90},index=['connection_ID']).set_index('connection_ID')
            cs=cs.append(cs_new)
            
cs.to_csv('connection_catalog_2.txt')
```
